```python
def readFASTA(inputfile):
    """Reads a FASTA sequence and retuns as string.Removes all the special characters"""
    with open(inputfile,"r")as f:
        seq = f.readline()
        seq = f.read()
        seq = seq.replace("\n", "")
        seq = seq.replace("\t", "")
    
    return seq
```
