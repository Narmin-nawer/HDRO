```python
import matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
#Fake data for plotting
labels = ("A", "T", "G", "C", "N")
sizes = [1857, 1184, 2889, 1267,1200]
```

```python
plt.pie(sizes, labels=labels)
plt.show()
```

```python
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue','orange']
```

```python
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors,)
plt.show()
```

```python
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, shadow=True)
plt.show()
```

# Explode

```python
explode = [0, 0, 0.1, 0, 0] 
#only explode the 3rd slice
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, shadow=True, startangle=90, explode=explode)
plt.legend(loc='best')
plt.tight_layout()
plt.savefig("pie_chart.pdf")
plt.show()
```

```python
explode = [0, 0, 0, 0, 0.1] 
#explode only the 5th slice 
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, shadow=True, startangle=90, explode=explode)
plt.legend(loc='best')
plt.tight_layout()
plt.savefig("pie_chart.pdf")
plt.show()
```
